```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow import keras
tf.random.set_seed(0)
```

```python
dataset=pd.read_csv('AABA_2006-01-01_to_2018-01-01.csv')
```

```python
dataset.head()
```

```python
scaler= MinMaxScaler((0,1))
dataset[["High"]]= scaler.fit_transform(dataset[["High"]])
data_test, data_train = [x for _, x in dataset.groupby(dataset['Date'] < '2017-01-01')]
data_test  = np.array(data_test.iloc[:,2])
data_train = np.array(data_train.iloc[:,2])
```

```python
X_train=[]
Y_train=[]
X_test=[]
Y_test=[]
for i in range(60,len(data_train)):
    X_train.append(data_train[i-60:i])
    Y_train.append(data_train[i])
for i in range(60,len(data_test)):
    X_test.append(data_test[i-60:i])
    Y_test.append(data_test[i])
x_train=np.array(X_train)
y_train=np.array(Y_train)
x_test=np.array(X_test)
y_test=np.array(Y_test)
```

```python
    def model1():
        inputs = keras.Input(shape=(60,1))
        x = layers.LSTM(units=100 , return_sequences=True)(inputs)
        x = layers.LSTM(units=100 , return_sequences=True)(x)
        x = layers.LSTM(units=100 , return_sequences=True)(x)
        x = layers.LSTM(units=100 , return_sequences=False)(x)
        x = layers.Dropout(0.2)(x)
        x = layers.Dense(32)(x)
        outputs = layers.Dense(1)(x)
        model = keras.Model(inputs=inputs, outputs=outputs)
        return model
```

```python
    def model2():
        inputs = keras.Input(shape=(60,1))
        x = layers.LSTM(units=50 , return_sequences=True )(inputs)
        x = layers.LSTM(units=50 , return_sequences=True)(x)
        x = layers.LSTM(units=50 , return_sequences=True)(x)
        x = layers.LSTM(units=50 , return_sequences=False)(x)
        x = layers.Dropout(0.2)(x)
        x = layers.Dense(64)(x)
        outputs = layers.Dense(1)(x)
        model = keras.Model(inputs=inputs, outputs=outputs)
        return model
```

```python
initial_learning_rate = 0.001
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate,
    decay_steps=1000000,
    decay_rate=0.99,
    staircase=False)
model1tmp = model1()
print(model1tmp.summary())
model1tmp.compile(
    loss="mean_squared_error",
    optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule)
)

model2tmp = model2()
print(model2tmp.summary())
model2tmp.compile(
    loss="mean_squared_error",
    optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule)
)
```

```python
history1=model1tmp.fit(x_train,y_train,epochs=10 , verbose=2 , batch_size=4,validation_data=(x_test,y_test) )
history2=model2tmp.fit(x_train,y_train,epochs=10 , verbose= 2,batch_size=4,validation_data=(x_test,y_test))
```

```python
pred1=model1tmp.predict(x_test)
pred1=scaler.inverse_transform(pred1)
y_test_temp=scaler.inverse_transform(y_test.reshape(-1,1))
pred2=model2tmp.predict(x_test)
pred2=scaler.inverse_transform(pred2)
plt.figure(figsize=(15, 8))
plt.title('Stock Prices History')
plt.plot(y_test_temp)
plt.plot(pred1)
plt.plot(pred2)
plt.xlabel('Date')
plt.legend(['Train', 'Prediction1','Prediction2'], loc='lower right')
plt.ylabel('Prices ($)')
mse1=(np.mean((pred1-y_test_temp)**2))
print(mse1)
mse2=(np.mean((pred2-y_test_temp)**2))
print(mse2)
```

```python
plt.figure(figsize=(15, 8))
plt.title('MSE')
plt.plot(history1.history["loss"])
plt.plot(history2.history["loss"])
plt.legend(['Model1_loss', 'Model2_loss'], loc='upper right')
```

```python
plt.plot(history1.history["val_loss"])
plt.plot(history2.history["val_loss"])
plt.legend(['Model1_val_loss', 'Model2_val_loss'], loc='upper right')
```
