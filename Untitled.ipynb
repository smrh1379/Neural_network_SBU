```python
import tensorflow_datasets as tfds
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import regularizers
tf.random.set_seed(1)
```

```python
classnames = {0: "rock", 1: "paper", 2: "scissors"}
x_train,y_train = tfds.as_numpy(
    tfds.load(name="rock_paper_scissors", split="train", as_supervised=True, batch_size=-1,shuffle_files=True))
test_ds = tfds.as_numpy(tfds.load(name="rock_paper_scissors", split="test", as_supervised=True, batch_size=-1))
```

```python
    for i in range(10):
        ax = plt.subplot(4, 3, i + 1)
        plt.imshow(x_train[i])
        plt.title(classnames.get(y_train[i]))
        plt.axis("off")
    plt.show()
```

```python
    def model1():
        inputs = keras.Input(shape=(300, 300, 3))
        x = layers.Rescaling(1. / 255)(inputs)
        x = layers.Conv2D(32, kernel_size=2, strides=2, padding="same", kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))(x)
        x = keras.activations.relu(x)
        x = layers.MaxPooling2D(padding="same")(x)
        x = layers.Conv2D(64, kernel_size=2, strides=2, padding="same",kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))(x)
        x = keras.activations.relu(x)
        x = layers.MaxPooling2D(padding="same")(x)
        x = layers.Conv2D(128, kernel_size=2, strides=2, padding="same",kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = keras.activations.relu(x)
        x = layers.MaxPooling2D(padding="same")(x)
        x = layers.Dropout(0.25)(x)
        x = layers.Flatten()(x)
        x = layers.Dense(32, activation="relu", kernel_initializer='he_uniform',kernel_regularizer=regularizers.l1_l2(0.001,0.01), bias_regularizer=regularizers.l1_l2(0.001,0.01))(x)
        x = layers.Dense(32, activation="relu", kernel_initializer='he_uniform',kernel_regularizer=regularizers.l1_l2(0.001,0.1), bias_regularizer=regularizers.l1_l2(0.001,0.1))(x)
        outputs = layers.Dense(3,activation="softmax")(x)
        model = keras.Model(inputs=inputs, outputs=outputs)
        return model
```

```python
    def model2():
        inputs = keras.Input(shape=(300, 300, 3))
        x = layers.Rescaling(1. / 255)(inputs)
        x = layers.Conv2D(8, kernel_size=2, strides=2, padding="same", kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))(x)
        x = keras.activations.relu(x)
        x = layers.MaxPooling2D(padding="same")(x)
        x = layers.Conv2D(32, kernel_size=2, strides=2, padding="same",kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))(x)
        x = keras.activations.relu(x)
        x = layers.MaxPooling2D(padding="same")(x)
        x = layers.Conv2D(64, kernel_size=2, strides=2, padding="same",kernel_regularizer=regularizers.l2(0.01), bias_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = keras.activations.relu(x)
        x = layers.MaxPooling2D(padding="same")(x)
        x = layers.Dropout(0.25)(x)
        x = layers.Flatten()(x)
        x = layers.Dense(32, activation="relu", kernel_initializer='he_uniform',kernel_regularizer=regularizers.l1_l2(0.001,0.01), bias_regularizer=regularizers.l1_l2(0.001,0.01))(x)
        x = layers.Dense(32, activation="relu", kernel_initializer='he_uniform',kernel_regularizer=regularizers.l1_l2(0.001,0.1), bias_regularizer=regularizers.l1_l2(0.001,0.1))(x)
        outputs = layers.Dense(3,activation="softmax")(x)
        model = keras.Model(inputs=inputs, outputs=outputs)
        return model
```

```python
    model1tmp = model1()
    print(model1tmp.summary())
    model1tmp.compile(
        loss=keras.losses.SparseCategoricalCrossentropy(),
        optimizer=keras.optimizers.Adam(),
        metrics=["accuracy"],
    )
```

```python
    ACCURACY_THRESHOLD=0.9
    class myCallback(tf.keras.callbacks.Callback):
        def on_epoch_end(self, epoch, logs={}):
            if (logs.get('val_accuracy') > ACCURACY_THRESHOLD and epoch>5):
                print("\nReached %2.2f%% accuracy, so stopping training!!" % (ACCURACY_THRESHOLD * 100))
                self.model.stop_training = True

    callbacks = myCallback()
```
